```python
plt.scatter(x,y)
plt.title("Google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (16,9) )
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5, linewidths = 10,
            edgecolors = "g")#verts = "<"
plt.title("Google play store apps scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```
