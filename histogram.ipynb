```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45,(100))
py_student_age = np.random.randint(15,40,(100))
```

```python
print(ml_students_age)
print(py_student_age)
```

```python
plt.hist(ml_students_age)
plt.title("Ml students age histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.show()
```

```python
# plt.hist(
#     x: 'ArrayLike | Sequence[ArrayLike]',
#     bins: 'int | Sequence[float] | str | None' = None,
#     range: 'tuple[float, float] | None' = None,
#     density: 'bool' = False,
#     weights: 'ArrayLike | None' = None,
#     cumulative: 'bool | float' = False,
#     bottom: 'ArrayLike | float | None' = None,
#     histtype: "Literal['bar', 'barstacked', 'step', 'stepfilled']" = 'bar',
#     align: "Literal['left', 'mid', 'right']" = 'mid',
#     orientation: "Literal['vertical', 'horizontal']" = 'vertical',
#     rwidth: 'float | None' = None,
#     log: 'bool' = False,
#     color: 'ColorType | Sequence[ColorType] | None' = None,
#     label: 'str | Sequence[str] | None' = None,
#     stacked: 'bool' = False,
#     *,
#     data=None,
#     **kwargs,
# ) -> 'tuple[np.ndarray | list[np.ndarray], np.ndarray, BarContainer | Polygon | list[BarContainer | Polygon]]'
# Docstring:
# Compute and plot a histogram.
```

```python
bins = [15,20,25,30,35,40,45]

plt.figure(figsize = (16,9))
plt.hist([ml_students_age,py_student_age], bins, rwidth = 0.8, histtype = "bar",
         orientation = "vertical", color = ["m","y"], label = ['ML Student','Py Student'])
# plt.hist(py_student_age, bins, rwidth = 0.8, histtype = "bar",
#          orientation = "vertical", color = "y", label = "Py Student")
plt.title("Ml students and py students histogram")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```
